```python
import pandas as pd
df = pd.read_csv('spaceship_tt.csv')
```

```python
df['CryoSleep'] = df['CryoSleep']*1
df['Transported'] = df['Transported']*1
df['VIP'] = df['VIP']*1
```

```python
df.isna().sum()
# 결측치를 가장 나이스하게 전처리했는가가 성능에 중요한 기준이 된다.
# 이진분류라 대부분 모델은 비슷하게 사용될 것
# 전처리가 너무나 중요하다.
# 모델링 했는데 성능이 안 나와요-> 전처리 다시 확인해야 한다.
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
#전처리 안 하고 베이스 모델로 정확도 평가해 보자
df_base=df[['CryoSleep','Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Transported']]
```

```python
df_base.dropna(inplace=True)
```

```python
df_base
```

```python
dt_clf = DecisionTreeClassifier(random_state=111)

X_train, X_test, y_train, y_test = train_test_split(df_base.drop('Transported',axis=1),df_base['Transported'],
                                                   test_size=0.3, random_state=111)

dt_clf = dt_clf.fit(X_train,y_train)
```

```python
X_train
```

```python
#실제 예측값 얼마인지?
from sklearn.metrics import accuracy_score
#하이퍼파라미터 튜닝하지 않은 base_model
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
print(dt_clf.get_params())
```

```python
df['psid_cnt']= df['PassengerId'].str[:4]
```
